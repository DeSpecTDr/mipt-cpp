```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
datadir = Path("data")
```

```python
def plot(name, title=None):
    path = datadir / name
    data = np.loadtxt(path)
    plt.title(name if title is None else title)
    plt.xlabel("N")
    plt.ylabel("T, ms")
    plt.scatter(data[:, 0], data[:, 1], label=path.stem, s=5)
```

# 1

```python
plot("baseline.txt")
plt.legend()
```

```python
plot("binary_baseline.txt")
```

```python
plot("binary_baseline.txt")
plt.xscale('log')
```
